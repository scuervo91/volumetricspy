```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from volumetricspy.stats import Spherical, Exponential, Gaussian
```

```python
s1 = Spherical(
    sill = 1.,
    range = 25,
    nuggets = 0.
)

s1.plot(np.linspace(0,50,50))
```

```python
s2 = Exponential(
    sill = 1.,
    range = 5,
    nuggets = 0.
)

s2.plot(np.linspace(0,50,50))
```

```python
s3 = Gaussian(
    sill = 1.,
    range = 15,
    nuggets = 0.
)

s3.plot(np.linspace(0,50,50))
```
